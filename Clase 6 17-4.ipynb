Calcular periodo fundamental de $cos(0.01  * \pi* n )$

$$T = \frac{2 \pi}{\omega}\hspace{1cm}
 T = \frac{2 \pi}{0.01 \pi}\hspace{1cm}
 T = \frac{2}{0.01}\hspace{1cm}
 T = 200$$


```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000  # frecuencia de muestreo de la señal en Hz
t = np.arange(-800, 800, 1)  # vector tiempo de 0 a 1 segundo
frecuencia_señal = 1000  # definimos una frecuencia de la señal senoidal de 10Hz
amplitud_señal = 1  # definimos la amplitud de la señal senoidal

señal_cosenoidal = amplitud_señal * np.cos(
    0.01 * np.pi  * t
)  # generamos la señal senoidal

plt.figure(figsize=(12, 4))
plt.plot(t, señal_cosenoidal)
plt.title("Señal Senoidal")
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Amplitud")

"""posiciones_x = [0.1, 0.4]
plt.vlines(
    x=posiciones_x,
    ymin=-amplitud_señal,
    ymax=amplitud_señal,
    colors="r",
    linestyles="--",
)
plt.xticks(posiciones_x, posiciones_x)
plt.axvlines()"""
plt.show()
```
