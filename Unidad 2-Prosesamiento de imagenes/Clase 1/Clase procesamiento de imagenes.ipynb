## Conceptos basicos de imagenes

Las imagenes corresponden a senales en 2D, cuyos ejes corresponden a filas y columnas. No todas las imagenes tienen las mismas caracteristicas, por ejemplo, pueden variar entre ellas el tamano y el color utilizado. 

En relacion con el tamano, la unidad de medida es el pixel, y la resolucion de la imagen esta dada por la cantidad de filas y columnas, asi una imagen de 100x100 tendra 10.000 pixeles de resolucion, mientras que una imagen de 1.000x1.000 tendra 1 Mpixeles de reslucion. Asi identificamos la primer diferencia entre las senales 1D y las imagenes, en el primer caso hablamos de muestras de la senal y ahora hablamos de pixeles de la imagen.

Otra caracteristica de la imagen corresponde a su color. Podemos encontrar imagenes en blanco y negro, a escala de grises y a color de tres bandas o canales. 

Las imagenes en blanco y negro (BW), tienen solamente un bit asociado a cada pixel de la imagen, de tal forma que, si la imagen tiene 1Mpixeles, entonces tendra 1Mbits. El valor de 1 corresponde al blanco, mientras que el valor de 0 corresponde al negro. 

Las imagenes a escala de grises tienen 8bits por cada pixel, y el rango de color va del negro (00000000) al blanco (11111111) pasando por distintas tonalidades de gris, para un total de 256 tonalidades. 

Finalmente tenemos las imagenes a color, que tipicamente se denominan RGB (Red, Green, Blue). En este caso por cada pixel de la imagen tenemos 8bits asociados a cada uno de los tres canales de color, para aun total de 24 bits por pixel.

Para ilustrar de mejor forma las diferencias de color de las imagenes estudiada, se presentan tres imagenes en BW, a escalas de grises y a color.

![Imagenes color](Imagenes_color.jpeg)

## Espacios de color

El espacio de color mas ampliamente conocido se denomina RGB, donde la imagen se representa en tres canales o bandas de color, una correspondiente al rojo, otra al verde y otra al azul. Cada color tiene 256 tonalidades distintas y en total se tiene una paleta de 16.776.216 colores. Veamos un ejemplo de imagen RGB donde se presentan las tres bandas de color.

![Espacio de color RGB](Espacio_de_color_RGB.jpeg)

Otro espacio de color corresponde a HSV (Hue, Saturation, Value) o HSB (Hue, Saturation, Brightness). En este espacio de color, la primera banda corresponde al tono de la imagen, la segunda a la saturacion de la imagen, y la tercera al brillo de la imagen. Veamos un ejemplo de imagen HSB.

![Espacio de color HSB](Espacio_de_color_HSB.jpeg)

El espacio de color HSB es ampliamente utilizado en tecnicas de reconocimiento y seguimiento de objetos por su mayor eficiencia, dado que permite una representación más precisa y detallada del color.

## Introduccion a la libreria OpenCV

Como introduccion a la libreria OpenCV en python veremos las operaciones basicas de *creacion o carga de imagenes desde un archivo y su visualizacion en pantalla*, *la obtencion de las caracteristicas de una imagen como sus dimensiones, numero de canales que la componen, tipo de dato con el que se almacena*, *creacion de una imagen a partir de una matriz de pixeles* y *modificacion del valor de los pixeles de una imagen*.

Antes de arrancar con el codigo en Python es conveniente presentar el sistema de coordenadas utilizado, en el que cada pixel se localiza en la imagen por su posicion horizontal y vertical (x,y). En la figura vemos que el origen (0,0) no esta en el centro de la imagen, sino en la parte superior izquierda, vemos que el pixel en color rojo estaria situado en la coordenada (3,1).

![Matriz imagen](matriz_imagen.jpg)

Otro concepto a tener en cuenta previo al uso de las funciones de OpenCV es el color, formado por las componente B, G, R, o lo que es lo mismo, los valores de la intensidad de sus componente azul, verde y rojo. Cualquier otro color se puede obtener a partir de una mezcla de ellos. El valor minimo de intensidad que pueden tener es 0 y el maximo 255.

Ejemplos de especificacion de colores utilizando la nomenclatura BGR seria:
- Blanco: (255,255,255)
- Negro: (0,0,0)
- Rojo: (0,0,255)
- Verde: (0,255,0)
- Azul: (255,0,0)
- Amarillo (0,255,255)
- Violeta: (255,0,255)
- Cian: (255,255,0)

Opcionalmente un pixel puede tener asociado un cuarto parametro que define su transparencia: el canal alfa. Sus valores van desde 0, complentamente transparente a 100, totalmente opaco.

#### Dividir imagen

```python
import cv2

img = cv2.imread("img/figuras.jpeg")

img_azul, img_verde, img_rojo = cv2.split(img)

cv2.imshow("Imagen azul", img_azul)
cv2.imshow("Imagen verde", img_verde)
cv2.imshow("Imagen roja", img_rojo)

# merge

img_original = cv2.merge([img_azul, img_verde, img_rojo])
cv2.imshow("Imagen recuperada", img_original)



cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Tarea: crear una grilla en verde
import numpy as np
import cv2

ancho_imagen = alto_imagen = 300
color_verde = (200, 255, 0)  # Verde (BGR)
color_rojo = (255, 0, 255)  # Verde (BGR)
espaciado = 99
vertical = 8
horizontal = 8

imagen = np.zeros((alto_imagen, ancho_imagen, 3), dtype=np.uint8)
imagen[:, :, :] = 255  # Color de fondo blanco (BGR)

for i in range(0, alto_imagen, espaciado):
    cv2.line(imagen, (0, i), (ancho_imagen - 1, i), color_verde, vertical)

for j in range(0, ancho_imagen, espaciado):
    cv2.line(imagen, (j, 0), (j, alto_imagen - 1), color_rojo, horizontal)


inicio_x = 0
inicio_y = 0
radio = 20
grosor = vertical
color_circulo = (255, 0, 255)  # Verde (BGR)
coordenadas_centro = ((inicio_x + espaciado) // 2, (inicio_y + espaciado) // 2)

cv2.circle(imagen, coordenadas_centro, radio, color_circulo, grosor)


cv2.imshow("Cuadricula", imagen)
cv2.waitKey(0)
```

#### Histograma

```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('img/obra.jpeg',0)
hist=cv2.calcHist([img],[0],None,[256],[0,256])

cv2.imshow('Imagen',img)

plt.xlabel('Nivel de luz')
plt.ylabel('Numero de pixeles')
plt.title('Histograma obra Moises')
plt.plot(hist)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread('img/obra.jpeg',0)

img_eq=cv2.equalizeHist(img)

hist_eq=cv2.calcHist([img_eq],[0],None,[256],[0,256])

cv2.imshow('Imagen ecualizada',img_eq)

plt.xlabel('Nivel de luz')
plt.ylabel('Numero de pixeles')
plt.title('Histograma equalizado')
plt.plot(hist_eq)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
