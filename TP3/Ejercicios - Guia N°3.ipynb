## Ejercicio 1 - Guía N°3: Muestreo de señales continuas

La señal $x_a(t) = \sin(2·\pi·100·t)$ se muestrea con un periodo de muestreo de $T_s = \frac{1}{400}$ segundos y se obtiene una señal en tiempo discreto $x[n]$. ¿Cuál es la señal resultante $x[n]$?

La señal muestreada en tiempo discreto $x(n)$ se obtiene evaluando la señal continua en los instantes de tiempo $t=n·T_s$, donde *n* es un número entero y $T_s$ es la tasa de muestreo. Así obtenemos: $$x(n) = x_a(n·T_s) = \sin(2·\pi·100·n·T_s) = \sin\left(\frac{2·\pi·100·n}{400}\right)= \sin\left(\frac{\pi·n}{2}\right)$$

Esta es nuestra señal discreta

Vamos ahora a graficar esta señal haciendo uso de las librerias matplotlib y numpy. Vamos a graficar la señal continua para un rango de tiempo continuo y la señal discreta en los puntos de muestreo.

```python
# Parámetros de la señal
f = 100  # Frecuencia de la señal (Hz)
fs = 400  # Frecuencia de muestreo (Hz)
Ts = 1/fs  # Periodo de muestreo (segundos)


# Definimos el vector tiempo para nuestra señal continua
tmin=0
tmax=0.05 #50 ms
t = np.linspace(tmin, tmax, 1000)  # definimos el vector tiempo de 50 ms
x_a = np.sin(2 * np.pi * f * t) # definimos la señal continua

"""
t: es un array de tiempo continuo desde 0 hasta 0.05 segundos (50 ms), tomando 1000 puntos en ese rango.
x_a: es la señal continua evaluada en los puntos de t
"""

# Definimos el vector de muestras para nuestra señal
nmin=0
nmax=0.05
n = np.arange(nmin, nmax, Ts)  # vector de muestras, mismo tiempo para la señal continua,
x_n = np.sin(2 * np.pi * f * n) # definimos la señal discreta

"""
n: es un array de instantes de tiempo discretos, desde 0 hasta 0.05 segundos, con incrementos de 𝑇s (nuestra tasa de muestreo)
x_n: es la señal muestreada evaluada en los puntos de n.
"""

# Graficamo la señal continua
plt.figure(figsize=(20, 6))
plt.plot(t, x_a)

# Graficamos la señal muestreada
plt.stem(n, x_n, 'r')

# Parámetros de la gráfica
plt.title('Señal Continua y Señal Muestreada')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud')
plt.grid(True)
```
