```python
import numpy as np
import matplotlib.pyplot as plt
```

Considere la siguiente señal sinusoidal analógica $$x_a(t) = 3*\sin(100*\pi * t)$$
1. Graficar la señal $x_a(t)$ para $0 < t \leq 30ms$. Determinar el periodo de la señal.

$2*\pi*f = \omega$ 
$f = \frac{100*\pi}{2*\pi} = 50$


```python
#determinamos el vector de tiempo
tmin=0
tmax=0.03
t= np.linspace(tmin,tmax,100)

#definimos los parametros de la señal
amp = 3
w=100*np.pi
f=w/(2*np.pi)
T= 1/f
print("El periodo de la señal es: ", T)

#definimos la señal
x_a=amp*np.sin(w*t)

#graficamos para el rango pedido

plt.plot(t,x_a)
```

 2. La señal $x_a(t)$ se muestrea con una tasa de muestreo $F_s = 300 \frac{muestras}{segundo}$.  Graficar la señal discreta en la misma ventana diagrama y determinar el periodo de muestreo de la señal en tiempo discreto.

El periodo de muestreo se calcula como la inversa de la frecuencia de muestreo: $T_s = \frac{1}{f_s} = \frac{1}{300} = 3.33ms$

La señal muestreada se expresa como (las variables t y n están relacionadas según la expresión: $ t = n·T_s = \frac{n}{F_s}$

$$x(n) = x_a(n·T_s)$$ $$x(n) = 3·\sin(100·\pi·n·T_S)$$ $$x(n) = 3·\sin(\frac{100}{300}·\pi·n) = 3·\sin(\frac{1}{3}·\pi·n)$$




```python
fs=300
Ts=1/fs

#definimos los parametros de la señal
amp = 3
w=100*np.pi
f=w/(2*np.pi)
T= 1/f

#definimos la señal continua
x_a=amp*np.sin(w*t)

#determinamos el vector de tiempo
tmin=0
tmax=0.03
t= np.linspace(tmin,tmax,100)

#definimos el vector de muestras

n=np.arange(0, 0.03,Ts)

print(n)
#definimos la señal discreta
x = 3*np.sin(100*np.pi*n)

print (x)

#graficamos
plt.plot(t,x_a)
plt.stem(n,x,'g')
```

Vemos que cada punto representa la muestra de la función en cada instante de tiempo.
