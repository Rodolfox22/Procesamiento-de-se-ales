```python
import numpy as np
import matplotlib.pyplot as plt


def acondicionar(valor, dig=0):
    if dig == 0:
        return int(valor)
    else:
        return round(valor, dig)


def filtrar(matriz):
    return list(set(matriz))


def periodo_t(vector_t, periodo):
    matriz = []
    for valor in vector_t:
        if round(abs(valor) % periodo, 2) == 0:
            valor = acondicionar(valor, 2)
            matriz.append(valor)
    return list(set(matriz))


# Cálculo de periodo fundamental por gáfico mostrando cruces por cero
def periodo_por_grafico(vector_t, señal, anterior=0, ajuste=0, redondeo_v=2, redondeo_t=1):
    matriz = []
    indice = 0
    indice_anterior = (
        -ajuste + anterior
    )  # Valor inicial para que detecte el primer cruce por cero
    indices = []

    # Evalúa si cruza por cero y filtra el resto de los cruces por el ajuste
    for valor in señal:
        if round(valor, redondeo_v) == 0.0 and indice - indice_anterior >= ajuste:
            matriz.append(round(vector_t[indice], redondeo_t))
            indices.append(indice)
            indice_anterior = indice
        indice += 1
    print(indices)
    print("El periodo fundamental calculado es: ", matriz[2] - matriz[1])
    return matriz
```

#### a) Calcular periodo fundamental de $cos(0.01  * \pi* n )$

$$T = \frac{2 \pi}{\omega}\hspace{1cm}
 T = \frac{2 \pi}{0.01 \pi}\hspace{1cm}
 T = \frac{2}{0.01}\hspace{1cm}
 T = 200$$


```python
w = 0.01 * np.pi
T = round(2 * np.pi / w, 2)
print("El período fundamental es: ", T)

t = np.arange(-800, 801, 1)

señal_cosenoidal = np.cos( w * t)  # generamos la señal coseno

posiciones_x = []

# Crear una figura y un eje
plt.figure(figsize=(15, 6))

# Trazar la señal coseno
plt.plot(t, señal_cosenoidal)

# Determinar las posiciones en el eje x donde trazar las líneas verticales
for valor in t:
    if valor % T == 0:
        posiciones_x.append(valor)
        plt.axvline(x=valor, color="red", linestyle="--")

# Establecer el título y las etiquetas de los ejes
plt.title("Señal Coseno")
plt.xticks(posiciones_x, posiciones_x)
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Amplitud")

# Mostrar el gráfico
plt.show()
```

#### e) $x(n) = cos( π* n/2) − sen( π* n/8) + 3 cos( π* n/4 + π /3)$

$$
f_{0} = \frac{\omega}{2 \pi}\hspace{1cm}
 
$$

Para calcular la frecuencia fundamental de la señal
$𝑥(𝑛)=cos(𝜋⋅𝑛/2)−sin(𝜋⋅𝑛/8)+3cos⁡(𝜋⋅𝑛/4+𝜋/3)$, necesitamos examinar cada término de la señal y determinar la frecuencia angular de cada uno. Luego, la frecuencia fundamental será el mínimo común múltiplo (m.c.m) de todas estas frecuencias angulares.<br>

Para cada término de la señal:<br>

**El primer término** <br>
$cos(\frac{𝜋⋅𝑛}{2})$ tiene una frecuencia angular de
$\hspace{0.5cm}\omega 1=\frac{𝜋}{2}$​.<br>
**El segundo término**<br>
$−sin⁡(\frac{𝜋⋅𝑛}{8})\hspace{0.5cm}$ tiene una frecuencia angular de
$\hspace{0.5cm}\omega 2=\frac {𝜋}8​$<br>
**El tercer término**<br>
$3cos⁡(\frac {𝜋⋅𝑛}4+\frac {𝜋}3)\hspace{0.5cm}$ tiene una frecuencia angular de
$\hspace{0.5cm}\omega 3=\frac {𝜋}4$​.<br>
El m.c.m de $\frac {𝜋}2$​ , $\frac {𝜋}8​$ y $\frac {𝜋}4$ es $𝜋$. Por lo tanto, la frecuencia fundamental de la señal es $\frac {1}𝜋​ Hz$.


```python
w = np.pi / 8
T = round(2 * np.pi / w, 2)
ff = round(w / (2 * np.pi), 2)
print("La frecuencia fundamental es: ", ff)
print("El periodo fundamental es: ", T)
rango = 60
t = np.arange(-rango, rango, 0.001)

# $x(n) = cos( π* n/2) − sen( π* n/8) + 3 cos( π* n/4 + π /3)$
señal = (
    np.cos(np.pi * t / 2)
    - np.sin(np.pi * t / 8)
    + 3 * np.cos(np.pi * t / 4 + np.pi / 3)
)  # generamos la señal coseno

posiciones_x = []

# Crear una figura y un eje
plt.figure(figsize=(15, 6))

# Dibuja la linea del eje x
# ceros = np.zeros((len(t), 1))
# plt.plot(t, ceros)

# Trazar la señal coseno
plt.plot(t, señal)


# Determinar las posiciones en el eje x donde trazar las líneas verticales
posiciones_x = periodo_t(t, T)
#posiciones_x = periodo_por_grafico(t, señal, anterior=0, ajuste=0)
for valor in posiciones_x:
    plt.axvline(x=valor, color="red", linestyle="--")

print(posiciones_x)


# Establecer el título y las etiquetas de los ejes
plt.title("Señal Senoidal")
plt.legend(["Frecuencia: {} Hz".format(ff)])
plt.xticks(posiciones_x, posiciones_x)
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Amplitud")

# Mostrar el gráfico
plt.show()
```
