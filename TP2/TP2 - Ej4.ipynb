```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la funciÃ³n x[n]
def fx(n):
    if n == -3 or n == -2 or n == -1:
        return 2
    elif n == 0:
        return 0
    elif n == 1:
        return 1
    elif n >= 2:
        return n

# Creamos un rango de valores para n
n = np.arange(-3, 5)
x=[]
# Calculamos los valores de x[n] para cada valor de n
for i in n:
    x.insert(i+3,fx(i))

plt.stem(n, x)

# Calcular los valores de x[-n] y x[-n+2]
#x1 = [fx(-i) for i in n]
#x2 = [fx(-i+2) for i in n]

```

```python
x1=[]
for i in n:
    x1.insert(i+3,fx(-i))

plt.stem(n, x1)

```

```python
x2=[]
for i in n:
    x2.insert(i+3,fx(-i+2))

plt.stem(n, x2)
```
