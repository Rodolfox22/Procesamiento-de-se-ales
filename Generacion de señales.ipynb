Para la resolución del ejercicio utilizaremos la biblioteca **numpy** para generar las señales y la biblioteca **matplotlib** para graficarlas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

También definimos una serie de parámetros comunes para generar y graficar estas señales, la frecuencia de muestreo (fs)  y el vector de tiempo (t).
La frecuencia de muestro es la cantidad de muestras que se toman por unidad de teimpo al digitalizar una señal analógica. Este es un parámetro fundamental a tener en cuenta en el proceso de muestreo y cuantización de una señal analógica para convertirla en una señal senoidal. Se expresa en muestras por segundo o en Hertz (Hz)

Cuando graficamos señales en python para poder visualizarlas necesitamos definir un vector tiempo que representa el dominio temporal de la señal. Este vector nos va a permitir especificar el intervalo de tiempo durante el cual se observa la señal y también determina la cantidad de puntos de muestra que se generan para representar la señal.

Por otro lado, la frecuencia de muestreo que definimos va a determinar cuantas muestras se toman por segundo para representar la señal. Si la frecuencia de muestreo es muy baja en comparación con la frecuencia de la señal, pueden ocurrir efectos no deseados como el aliasing.

```python
fs = 1000 # frecuencia de muestreo de la señal en Hz
t = np.arange(0, 1, 1/fs) # vector tiempo de 0 a 1 segundo

#En este ejemplo definimos una frecuencia de muestreo de 1000Hz, es decir que tomamos 1000 muestras por segundo de la señal, 
#el vector tiempo va de 0 a 1 segundo con un intervalo de tiempo entre muestras igual a 1/fs
```

### Señal senoidal

```python
frecuencia_señal = 10 # definimos una frecuencia de la señal senoidal de 10Hz
amplitud_señal = 1 # definimos la amplitud de la señal senoidal

señal_senoidal = amplitud_señal * np.sin(2*np.pi*frecuencia_señal*t) # generamos la señal senoidal

plt.plot(t, señal_senoidal)
plt.title('Señal Senoidal')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
plt.show()
```

### Señal sinc

```python
señal_sinc = np.sinc(10*t) # utlizamos la función sinc de numpy para generar la señal

#Graficamos la señal
plt.plot(t, señal_sinc)
plt.title('Señal sinc')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
plt.show()
```

### Señal cuadrada

```python
from scipy import signal

#señal_cuadrada = amplitud_señal * np.sign(np.sin(2*np.pi * frecuencia_señal * t))
señal_cuadrada = amplitud_señal * signal.square(2 * np.pi * frecuencia_señal * t)

#Graficamos la señal cuadrada
plt.plot(t, señal_cuadrada)
plt.title('Señal cuadrada')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
plt.show()
```

### Señal triangular

```python
señal_triangular = amplitud_señal * np.abs((2 * (t * frecuencia_señal) % 2) - 1) - amplitud_señal/2

#Graficamos la señal triangular
plt.plot(t, señal_triangular)
plt.title('Señal triangular')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
plt.show()
```

### Delta de Dirac

```python
delta_dirac = np.zeros_like(t)
delta_dirac[500] = 1  # Impulso en el instante de tiempo 500

# Graficar el delta de Dirac
plt.plot(t, delta_dirac)
plt.title('Delta de Dirac')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Ruido aleatorio

```python
ruido_aleatorio = np.random.normal(0, 1, len(t))  # Ruido gaussiano

# Graficar el ruido aleatorio
plt.plot(t, ruido_aleatorio)
plt.title('Ruido Aleatorio')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```
