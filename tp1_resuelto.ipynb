## Ejercicio 3

Escribí un programa que solicite al usuario el ingreso de una temperatura en escala Fahrenheit (debe permitir decimales) y le muestre el equivalente en grados Celsius. La fórmula de conversión que se usa para este cálculo es: ° Celsius = (5/9)*(° Fahrenheit - 32)

```python
# Crear un objeto de la clase UnitRegistry
ureg = UnitRegistry()

# Solicitar al usuario la temperatura en Fahrenheit
temp_fahrenheit = float(input("Ingresa la temperatura en Fahrenheit: "))

# Convertir la temperatura a la unidad 'degree_Fahrenheit' utilizando pint
temp_fahrenheit_quantity = ureg.Quantity(temp_fahrenheit, 'degree_Fahrenheit')
print(temp_fahrenheit_quantity)

# Realizar la conversión a Celsius utilizando la fórmula proporcionada
temp_celsius_quantity = (5/9) * (temp_fahrenheit_quantity - ureg.Quantity(32, 'degree_Fahrenheit'))

print("temperatura en celsius: ",temp_celsius_quantity*((ureg.Quantity(1,'degree_Celsius'))/(ureg.Quantity(1,'degree_Fahrenheit'))))
# Imprimir el resultado
print(f"La temperatura equivalente en grados Celsius es: {temp_fahrenheit_quantity.to(ureg.degree_Celsius)}")
```

# Ejercicio 8

Escribí un programa que, dado un número entero positivo, calcule y muestre su factorial. El factorial de un número se obtiene multiplicando todos los números enteros positivos que hay entre el 1 y ese número. El factorial de 0 es 1.

```python
def calcular_factorial(numero):
    # Inicializamos el factorial como 1
    factorial = 1

    # Verificamos si el número es 0, ya que el factorial es 1
    if numero == 0:
        return factorial

    # Para numero mayores que 0 calculamos el factorial. Utilizamos el bucle for y la funció in range
    for i in range(1, numero + 1):
        # Multiplicamos el factorial por el número actual en la iteración
        factorial = factorial * i

    return factorial


# Solicitamos al usuario que ingrese un número entero positivo
numero = int(input("Ingresa un número entero positivo: "))

# Verificamos si el número ingresado es positivo
if numero < 0:
    print("El número ingresado no es positivo.")
else:
    # Calculamos el factorial del número ingresado
    resultado = calcular_factorial(numero)
    print("El factorial de", numero, "es:", resultado)
```

Para resolver este ejercicio utilizamos un bucle **for** y la función **range**.
La función range() genera una secuencia de números comenzando desde el primer argumento (el iterador inicial) hasta el segundo argumento menos 1 (iterador final)
El bucle **for in** indica que el for va a iterar sobre cada valor de la secuencia generada por la función range. En cada iteración el valor actual se asigna a la variable i.

Luego escribe una función que se llame "media" que tome una lista y devuelva una nueva lista que contenga todos los elementos de la lista anterior menos el primero y el último.

```python
def elimina (lista):
    if len(lista)>=2:
        del lista[0]
        del lista[-1]
        return(lista)
    else:
        return []

def media (lista):
    if len(lista)>=2:
        return(lista[1:-1])
    else:
        return([])

lista_eliminar = [1,2,3,4]
lista_media = [5,6,7,8]
resultado_elimina = elimina(lista_eliminar)
print(resultado_elimina)
print(lista_eliminar)

resultado_media = media(lista_media)
print(resultado_media)
print(lista_media)
```
