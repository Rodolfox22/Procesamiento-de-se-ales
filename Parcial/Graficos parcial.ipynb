```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fx6(n):
    if n < -7 and n >= 7:
        return 0
    elif n >= -7 and n < 1:
        return n + 4
    elif n >= 1 and n < 7:
        return 4
    else:
        return 0


n = np.arange(-10, 12)
x = []

# Creo la matriz x a partir de los valores n
for i in n:
    x.insert(i + 10, fx6(i))

plt.figure(figsize=(10, 4))
plt.axvline(x=0, color="red", linestyle="--")
plt.stem(n, x)
plt.title("X(n)")
print(n)
print(x)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
f_signal = 5000  # Frecuencia de la señal en Hz
f_sampling = 8000  # Frecuencia de muestreo en Hz
t_end = 0.002  # Duración de la señal en segundos

# Tiempo continuo
t_cont = np.linspace(0, t_end, int(f_sampling * t_end * 100))  # Más puntos para la señal continua
signal_cont = np.sin(2 * np.pi * f_signal * t_cont)

# Tiempo muestreado
t_sampled = np.arange(0, t_end, 1/f_sampling)
signal_sampled = np.sin(2 * np.pi * f_signal * t_sampled)

# Graficar
plt.figure(figsize=(10, 6))

# Señal continua
plt.plot(t_cont, signal_cont, label='Señal continua (5 kHz)', alpha=0.7)

# Señal muestreada
plt.stem(t_sampled, signal_sampled, linefmt='r-', markerfmt='ro', basefmt='r-', label='Señal muestreada (8 kHz)', use_line_collection=True)

# Etiquetas y leyenda
plt.title('Señal de 5 kHz muestreada a 8 kHz')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
