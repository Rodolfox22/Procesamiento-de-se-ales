# Desplazamiento de se침ales

Una se침al $x(t)$ que se retrasa **k** segundos se representa como una versi칩n desplazada hacia la derecha del eje **t**. Esto es: $$\phi(t+k) = x(t)$$ o escrito de otra forma: $$\phi(t) = x(t-k)$$

Podemos ver que el signo determina el desplazamiento:
- hacia la derecha si se resta **k**
- hacia la izquierda si se suma **k**

Veamos un ejemplo con la funci칩n $x(t) = \sin(\omega * t)$

$x(t) = \sin(\omega_0*t)$ donde el periodo es: $T = \frac{2*\pi}{\omega_0}$, tiene unidades de segundo

```python
# Definimos los par치metros de la funci칩n
f=1
w=1
T=2*np.pi/w

k=np.pi/2

#Definimos el vector tiempo
ti=0
tf=2*T

fs=1000

t=np.arange(ti,tf,1/fs)
#print(t)


seno = np.sin(w*t)
seno1 = np.sin(w*t-k) #Desplazamiento hacia la derecha
seno2 = np.sin(w*t+k) #Desplazamiento hacia la izquierda

print(np.sin(0))
print(np.sin(np.pi/2))
print(np.sin(np.pi))
print(np.sin(np.pi*1.5))
#print(np.sin(2*np.pi)) revisar


plt.subplot(3,1,1)
plt.plot(t,seno)
for i in range(ti, 3):
    plt.axvline(x=i * T, color='r', linestyle='--', linewidth=1)
plt.subplot(3,1,2)
plt.plot(t,seno1)
for i in range(ti, 3):
    plt.axvline(x=i * T, color='r', linestyle='--', linewidth=1)


plt.subplot(3,1,3)
plt.plot(t,seno2)
for i in range(ti, 3):
    plt.axvline(x=i * T, color='r', linestyle='--', linewidth=1)

plt.figure(2)
plt.plot(t,seno, color='red')
plt.plot(t, seno1, color='blue')
plt.plot(t, seno2, color='green')

```

## Escalamiento de se침ales

La compresi칩n o expansi칩n de la se침al en el tiempo se denomina **escalamiento** en el tiempo. Por ejemplo, consideremos una se침al $x(t)$ afecta en el tiempo por un factor de 2. Tenemos que: $$\phi \left(\frac{t}{2}\right) = x(t)$$ o lo que es lo mismo: $$ \phi (t) = x(2*t)$$.
Trabajemos nuevamente con la se침al $x(t) = \sin(t)$

```python
#Definimos los parametros de la se침al
w=1
T=2*np.pi

k=np.pi/2

ti=0
tf=2*T

#Definimos el vector tiempo

fs=1000

t=np.arange(ti,tf,1/fs)

seno = np.sin(w*t)
seno2 = np.sin(w*0.5*t)
seno3 = np.sin(w*2*t)

plt.subplot(3,1,1)
plt.plot(t,seno)

plt.subplot(3,1,2)
plt.plot(t,seno2)

plt.subplot(3,1,3)
plt.plot(t,seno3)

plt.figure(2)
plt.plot(t,seno, color='red')
plt.plot(t, seno2, color='blue')
plt.plot(t, seno3, color='green')
```

Resumiendo, el efecto de transformar la variable independiente de una se침al $x(t)$ para obtener la se침al modificada de la forma $$x(a*t+b)$$
Con la transformaci칩n, la variable independiente conserva la forma de $x(t)$. La se침al puede ser:
- Alargada linealmente cuando $|a|<1$
- Comprimida si $|a|>1$
- invertida en el teimpo si $a<0$
- desplazada en el teimpo si b es diferente de cero:
    - siendo desplazada a la derecha si se resta el valor de $|b|$
    - siendo desplazada a la izquierda si se suma el valor de $|b|$
