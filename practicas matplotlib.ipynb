```python
import matplotlib.pyplot as plt

# Datos
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 3, 11]

# Crear el gráfico
plt.plot(x, y)

# Agregar etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de ejemplo')

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Crear una matriz de ejemplo
matriz = np.random.rand(10, 10)
# Mostrar la matriz
plt.imshow(matriz, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Agregar una barra de color para referencia
plt.title('Matriz de ejemplo')
plt.show()
```

```python
import numpy as np

# Crear una matriz unidimensional con arange()
matriz_unidimensional = np.arange(2, 370, step=9 )  # crea una secuencia del 1 al 12

# Calcular cuántos elementos faltan para completar las 3 filas
elementos_faltantes = matriz_unidimensional.size % 3 
falta = [0,2,1]

# Agregar ceros a la matriz unidimensional
matriz_unidimensional = np.pad(matriz_unidimensional, (0, falta[elementos_faltantes]), mode='constant')

# Dar forma a la matriz unidimensional en una matriz bidimensional
matriz_bidimensional = matriz_unidimensional.reshape(3, -1)  # matriz de 3 filas y 4 columnas

print("Matriz bidimensional:")
print(matriz_bidimensional)
```
