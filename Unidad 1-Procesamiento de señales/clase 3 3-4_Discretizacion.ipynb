```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
frecuencia_original = 5
t = np.linspace(0, 1, 500)
amplitud_senhal = 1

senhal_senoidal = amplitud_senhal * np.sin(2 * np.pi * frecuencia_original * t)

plt.plot(t, senhal_senoidal)

# Frecuencias de muestreo: 1000; 100; 25; 10; 4; 1
conteo = 0

muestreo = [100, 50, 40, 17, 7, 3]
fs = muestreo[0]
t1 = np.arange(0, 1, 1 / fs)
senhal_discreta = np.sin(2 * np.pi * frecuencia_original * t1)

plt.subplot(2, 1, 1)
plt.title(f"Muestreo: {fs}")
plt.plot(t, senhal_senoidal)
plt.subplot(2, 1, 2)
plt.stem(t1, senhal_discreta)
```
