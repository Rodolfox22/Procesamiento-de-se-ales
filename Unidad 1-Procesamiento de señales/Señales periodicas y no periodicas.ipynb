# SEÑALES PERIÓDICAS Y NO PERIÓDICAS

Una característica fundamental que distingue algunas señales es su propiedad de repetirse de manera exacta cada cierto tiempo-

#### *Una señal cuyos valores se repiten de manera exacta cada T unidades de tiempo se denomina **señal periódica***

Así, una señal continua cuya función es $f\left(t\right)$, será periódica de periodo *T* si exsite una constante $T>0$ tal que $$f\left(t\right) = f\left(t+T\right)$$ para todo *t*

El valor mínimo (positivo) de la constante *T* par ael cual se cumple lo anterior se denomina **periodo fundamental** de la señal $f\left(t\right)$

##### Veamos algunos ejemplos

**¿Cuál es el periodo fundamental de las siguientes señales?**

1. $f\left(t\right) = sin(\omega t)$


Dado que $$sin(t) = sin(t + 2k\pi)$$ para $k = 0,\pm1,\pm2,...$ 

Entonces $$\sin(\omega t) = \sin(\omega t + 2 k \pi) = \sin\left(\omega\left(t+\frac{2 k \pi}{\omega} \right) \right) = f\left(t + \frac{2 k \pi}{\omega} \right)  $$ para $k = 0,\pm1,\pm2,...$

Por lo tanto para *k=1* se tiene que $T = \frac{2\pi}{\omega}$ radianes.

```python
A=1
w = 2*np.pi*f
T = 2*np.pi
f = 1/T

k=2

fs = 1000
ti=0
tf=ti+k*T
t = np.arange(ti, tf, 1/fs)

seno = A * np.sin(t + k*T/w) # generamos la señal senoidal
#print(señal_senoidal)
plt.plot(t, seno)
plt.title('Señal Senoidal')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
for i in range(ti, 3):
    plt.axvline(x=i * T, color='r', linestyle='--', linewidth=1)
plt.show()
```

2. $f(t) = \sin\left(\frac{t}{3}\right) + cos\left(\frac{t}{4}\right)$

De la definición tenemos: $$f(t) = f(t+T) = \sin\left(\frac{t+T}{3}\right) + \cos\left(\frac{t+T}{4}\right)$$

Por otro lado, $$f(t) = \sin\left(\frac{t}{3} + 2k\pi\right) + \cos\left(\frac{t}{4} + 2k\pi\right)$$, para $k=0,\pm1,\pm2,...$
Es decir, para que se cumpla lo anterior se requiere que $$\frac{T}{3} = 2k_1\pi$$ $$\frac{T}{4} = 2k_2\pi$$

de donde obtenemos $$T = 6k_1\pi = 8k_2\pi$$

por lo tanto los valores mínimos para los cuales se cumple lo anterior son $k_1=4$ y $k_2=3$. es decir $T=24\pi$ *radianes*

```python
A=1
T=24*np.pi
f=1/T
k=2

fs = 1000
ti=0
tf=ti + k*T
t = np.linspace(ti, tf, fs)

seno = A*np.sin(t/3)
coseno = A*np.cos(t/4)
señal=seno+coseno

plt.plot(t, señal)
plt.title('Señal')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
for i in range(ti, 3):
    plt.axvline(x=i * T, color='r', linestyle='--', linewidth=1)
plt.show()            
```
