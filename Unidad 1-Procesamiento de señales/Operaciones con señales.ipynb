```python
import numpy as np
import matplotlib.pyplot as plt
```

# Operaciones con señales

Realice las siguientes operaciones básicas sobre una señal senoidal:

1. expansión
2. compresión
3. inversión
4. rectificación
5. cuantización en 8 niveles
6. traslación


```python
# Definimos una señal senoidal
fs = 1000 # frecuencia de muestreo de la señal en Hz
t = np.linspace(0, 1, 1000) # vector tiempo de 0 a 1 segundo
frecuencia_señal = 10 # definimos una frecuencia de la señal senoidal de 10Hz
amplitud_señal = 1 # definimos la amplitud de la señal senoidal

señal_senoidal = amplitud_señal * np.sin(2*np.pi*frecuencia_señal*t) # generamos la señal senoidal

plt.plot(t, señal_senoidal)
plt.title('Señal Senoidal')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Amplitud')
plt.show()
```

### Expansión

La expansión es una operación de dominio. Implica aumentar la duración de la señal. Esto se logra escalando en el eje del tiempo, es decir, para cada punto en el eje del tiempo (eje x) se lo multiplica por un factor mayor que 1. Por ejemplo, si tenemos una señal con una duración de 1 segundo y aplicamos un factor de expansión de 2, la señal resultante tendrá ahora una duración de 2 segundos. Vemos una disminución de la frecuencia de la señal

```python
factor_expansion = 2  # Factor de expansión
nuevo_t = t * factor_expansion
senoidal_expandida = np.interp(t, nuevo_t, señal_senoidal)

# Graficamos la señal expandida
plt.plot(t, senoidal_expandida)
plt.title('Señal Senoidal Expandida')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Compresión

La compresión en el dominio del teimpo implica reducir la duración de una señal, a diferencia de la expansión, ahora para cada punto en el eje del tiempo, lo multiplicamos por un valor entre 0 y 1. Así, si tenemos una señal senoidal con una duración de 1 segundo, al aplicar un factor de compresión de 0.5, la señal resultante tendrá una duración de 0.5 segundos. Vemos un aumento en la frecuencia de la señal.

```python
factor_compresion = 0.5  # Factor de expansión
nuevo_t = t * factor_compresion
senoidal_comprimida = np.interp(t, nuevo_t, señal_senoidal)

# Graficamos la señal comprimida
plt.plot(t, senoidal_comprimida)
plt.title('Señal Senoidal Comprimida')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Inversión

La inversión implica invertir el orden de los puntos de la señal en el eje del tiempo, es decir reflejar la señal respecto al eje vertical. Por ejemplo, si tenemos una señal que va desde el tiempo 0 hasta 1 segundo, al realizar la inversión, la señal irá ahora desde 1 hasta 0.

```python
senoidal_invertida = señal_senoidal[::-1]

# Graficamos la señal comprimida
plt.plot(t, senoidal_invertida)
plt.title('Señal Senoidal Invertida')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Rectificación

La rectificación de una señal implica hacer que todos sus valores sean no negativos, es decir que cualquier parte de la señal que esté por debajo del eje horizontal se refleja sobre el eje para que la señal resultante sea siempre positiva o cero.

```python
senoidal_rectificada = np.abs(señal_senoidal)

# Graficamos la señal rectificada
plt.plot(t, senoidal_rectificada)
plt.title('Señal Senoidal Rectificada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Cuantización

La cuantización es el proceso de reducir el número de valores posibles de una señal a un conjunto discreto y finito de valores. 

```python
niveles = 8
amplitud_max = np.max(señal_senoidal)
amplitud_min = np.min(señal_senoidal)
paso = (amplitud_max - amplitud_min) / niveles
senoidal_cuantizada = np.round(señal_senoidal / paso) * paso

# Graficamos la señal cuantizada
plt.plot(t, senoidal_cuantizada)
plt.title('Señal Senoidal Cuantizada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()
```

### Traslación

La traslación de una señal implica desplazar toda la señal a lo largo del eje del tiempo, es decir, mover la señal hacia la izquierda o hacia la derecha en el tiempo. Esto se logra desplazando cada punto de la señal a lo largo del eje del tiempo por una cierta cantidad de unidades de tiempo. 

```python
desplazamiento=500
senoidal_trasladada = np.roll(señal_senoidal, desplazamiento)
if desplazamiento > 0:
    senoidal_trasladada[:desplazamiento] = 0
elif desplazamiento <0:
    senoidal_trasladada[desplazamiento:] = 0

plt.plot(t, senoidal_trasladada)
plt.title('Señal Trasladada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
```
