# Generacion de señales

Genere y grafique las siguientes señales
1. Señal seniodal
2. Sinc
3. Onda cuadrada
3. Delta de Dirac
3. Ruido aleatorio

```python
import numpy as np
import matplotlib.pyplot as plt

frecuencia_muestreo = 0
vector_de_tiempo = 0 # existen 2 formas de hacerlo

# Frecuencia de muestreo
fs = 1000 # teorema de Nyquist, la fecuencia tiene que ser 2 veces la señal que nosotros queremos muestrear

# Vector de tiempo
t1 = np.arange(0, 1, 1/fs)

# Vector de tiempo, metodo 2
t2 = np.linspace(0 ,1 ,fs )

#print(t1)
#print(t2)

```

## 1. Señal Senoidal

```python
frecuencia_senhal = 5
amplitud_senhal = 1

senhal_senoidal = amplitud_senhal * np.sin(2 * np.pi * frecuencia_senhal * t1)
# print(senhal_senoidal)

plt.plot(t1, senhal_senoidal)
```

## 2. Sinc: seno complementario

```python
#t2 = [value for index, value in enumerate(t1) if index % 50 == 0] # Forma personal de modificar el muestreo de la señal

senhal_sinc =np.sinc(15 * t1)

plt.plot(t1, senhal_sinc)
plt.title("Función Sinc")
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Amplitud")
```

## 3. Onda cuadrada

```python
# pip install scipy
# from scipy import signal
# senhal_cuadrada = amplitud_senhal * signal.square(2*np.pi*frecuencia_senhal*t1)

senhal_cuadrada = amplitud_senhal * np.sign(np.sin(2 * np.pi * frecuencia_senhal * t1))

plt.subplot(1,2,1)
plt.plot(t1, senhal_cuadrada)
plt.subplot(1,2,2)
plt.stem(t1, senhal_cuadrada)
```

## 4. Delta de Dirac

```python
delta_dirac = np.zeros_like(t1)
delta_dirac[500] = 1

plt.plot(t1, delta_dirac)
```

## 5. Ruido aleatorio, señal estocastica

```python
ruido_aleatorio = np.random.normal(0, 1, len(t1))

plt.plot(t1, ruido_aleatorio)
```

## 6. Diente de sierra

```python
# Parámetros
amplitud = 1.0  # Amplitud de la onda de diente de sierra
frecuencia = 1.0  # Frecuencia de la onda de diente de sierra
periodo = 1.0 / frecuencia  # Periodo de la onda de diente de sierra
muestras_por_periodo = 100  # Número de muestras por período
num_periodos = 5  # Número de periodos que quieres generar

# Calcular los puntos de la onda de diente de sierra
t = np.linspace(0, num_periodos * periodo, num_periodos * muestras_por_periodo)  # Tiempo
onda_diente_de_sierra = amplitud * (2 * (t / periodo - np.floor(0.5 + t / periodo)))

# Graficar la onda de diente de sierra
plt.plot(t, onda_diente_de_sierra)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Onda de diente de sierra')
plt.grid(True)
plt.show()
```
